```python
import math

def sigmoid(z):
  return 1.0 / (1.0 + math.exp(-z))

def sigmoid_derivative(sigmoid_output):
  return sigmoid_output * (1.0 - sigmoid_output)

x1, x2 = 1.0, 2.0

y = 1.0

w1 = 0.10
w2 = -0.30
w3 = 0.50

alpha = 100.0
num_epochs = 1000

for epoch in range(num_epochs):
  z1 = w1 * x1 + w2 * x2
  h1 = sigmoid(z1)

  z2 = w3 * h1
  y_hat = sigmoid(z2)

  # Loss calculation, implementing the MSE Loss
  error = 0.5 * (y - y_hat)**2

  # Backward Pass (Using the chain rule)
  dL_dy = y_hat - y

  dy_dz2 = sigmoid_derivative(y_hat)

  dL_dz2 = dL_dy * dy_dz2

  dL_dw3 = dL_dz2 * h1

  dh1_dz1 = sigmoid_derivative(h1)

  dL_dz1 = dL_dz2 * w3 * dh1_dz1

  dL_dw1 = dL_dz1 * x1

  dL_dw2 = dL_dz1 * x2

  # Gradient descent
  w1 -= alpha * dL_dw1

  w2 -= alpha * dL_dw2

  w3 -= alpha * dL_dw3

  print(f'Epoch {epoch + 1}: ')
  print(f'y_hat = {y_hat:.8f}')
  print(f'Error = {error: .8f}')
  print(f'Updated weights: w1={w1:.8f}, w2={w2:.8f}, w3={w3:.8f}')

print('Training completed!')
```
